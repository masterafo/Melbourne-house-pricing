```python
import pandas as pd
```

```python
melb_data = pd.read_csv('melb_data.csv/melb_data.csv')
```

```python
y = melb_data.Price
```

```python
melb_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
x = melb_data[melb_features]
x.describe()
```

```python
x.head()
```

```python
x.shape
missing_val_count_by_column = (x.isnull().sum())
print(missing_val_count_by_column)
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
#from sklearn.ensemble import RandomForestRegressor
#melb_model = RandomForestRegressor(n_estimators=50, random_state=0)
#melb_model.fit(x_train, y_train)
```

```python
from xgboost import XGBRegressor
melb_model = XGBRegressor(n_estimators=500)
melb_model.fit(x_train, y_train)
```

```python
melb_test = melb_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, melb_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, melb_test)
```
